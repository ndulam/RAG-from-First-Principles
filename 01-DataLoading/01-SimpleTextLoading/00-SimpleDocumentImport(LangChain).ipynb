## Load a document and create Documents with LangChain

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader("/home/huangj2/Documents/20250202_RAG_Book/data/BlackMythWukong/black_myth_wukong_setup.txt")
documents = loader.load()
print(documents)
```

## Create Document objects

```python
from langchain_core.documents import Document
documents = [
    Document(
        page_content="Wukong is the eldest disciple.",
        metadata={"source": "master_and_disciples.txt"},
    ),
    Document(
        page_content="Bajie is the second disciple.",
        metadata={"source": "master_and_disciples.txt "},
    ),
]
print(documents)
```

## Load all documents in a directory

```python
from langchain_community.document_loaders import DirectoryLoader
loader = DirectoryLoader("/home/huangj2/Documents/20250202_RAG_Book/data/BlackMythWukong")
docs = loader.load()
print(f"Number of documents: {len(docs)}")  # print the total document count
print(docs[0])  # print the first document
```

## Specify loading parameters

```python
from langchain_community.document_loaders import DirectoryLoader
loader = DirectoryLoader("/home/huangj2/Documents/20250202_RAG_Book/data/BlackMythWukong",
                         # glob="**/*.md",
                         use_multithreading=True,
                         show_progress=True,
                         )
docs = loader.load()
print(f"Number of documents: {len(docs)}")  # print the total document count
print(docs[0])  # print the first document
```

## Change the loading tool

```python
from langchain_community.document_loaders import DirectoryLoader, TextLoader
# Load every Markdown file in the directory
loader = DirectoryLoader("/home/huangj2/Documents/20250202_RAG_Book/data/BlackMythWukong",
                         glob="**/*.md",
                         loader_cls=TextLoader)
docs = loader.load()
print(docs[0].page_content[:100])  # print the first 100 characters of the first document's content
```

## Skip files that error out

```python
from langchain_community.document_loaders import DirectoryLoader, TextLoader
# Load every file in the directory, skipping files that error out, since some files are images that TextLoader cannot load
loader = DirectoryLoader("/home/huangj2/Documents/20250202_RAG_Book/data/BlackMythWukong",
                          silent_errors=True,
                         loader_cls=TextLoader)

docs = loader.load()
print(docs[0].page_content[:100])  # print the first 100 characters of the first document's content
```
